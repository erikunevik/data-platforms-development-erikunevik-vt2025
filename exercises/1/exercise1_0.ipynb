```python

import pandas as pd
import matplotlib.pyplot as plt

df_svenska_samtliga = df_svenska[df_svenska['Huvudman'] == 'Samtliga']
df_svenska_andra_samtliga = df_svenska_andra[df_svenska_andra['Huvudman'] == 'Samtliga']
df_matematik_samtliga = df_matematik[df_matematik['Huvudman'] == 'Samtliga']
df_engelska_samtliga= df_engelska[df_engelska['Huvudman'] == 'Samtliga']


subjects = ["Svenska", "Svenska som andraspråk", "Matematik", "Engelska", ]

scores = [
    df_svenska_samtliga["Totalt (poäng)"].sum(),
    df_svenska_andra_samtliga["Totalt (poäng)"].sum(),
    df_matematik_samtliga["Totalt (poäng)"].sum(),
    df_engelska_samtliga["Totalt (poäng)"].sum()
]

data = pd.DataFrame({"Subject": subjects, "Score": scores})

plt.figure(figsize=(8, 6))
plt.bar(data["Subject"], data["Score"], width = 0.25, color=["blue"])
plt.ylim(bottom = 5)
plt.title("Totalpoäng baserat på ämne")
plt.xlabel("Ämne")
plt.ylabel("Poäng")
plt.show()
    
    
```
