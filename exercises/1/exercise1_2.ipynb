2 E) Find out the five sports with highest median age

```python
high_median_sql = query("""
    SELECT Sport, MEDIAN(Age) AS Median_Age
    FROM df
    GROUP BY Sport
    ORDER BY Median_Age DESC
    LIMIT 5
""")

print(high_median_sql)
```

```python
high_median_df = (df.groupby("Sport")["Age"].median().sort_values(ascending=False).head(5))  

print(high_median_df)
```

1 F) Find out the five sports with lowest median age

```python
low_median_sql = query("""
            SELECT Sport, MEDIAN(Age)
            FROM df
            GROUP BY Sport
            ORDER BY MEDIAN(Age)
            LIMIT 5
            """)
print(low_median_sql)
          
        
```

```python
low_median_df = (df.groupby("Sport")["Age"].median().sort_values(ascending = False).head(5))

print(low_median_df)
```

2 F) Find out top 10 countries after number of gold medals

```python
df.columns
```

```python
print(df['Medal'].unique())  # This will show all unique values in the column.
```

```python
gold_sql = query("""
    SELECT Team, COUNT(Medal) AS Gold_Medal_Count
    FROM df
    WHERE Medal = 'Gold' AND Medal IS NOT NULL
    GROUP BY Team
    ORDER BY Gold_Medal_Count DESC
    LIMIT 10
""")
print(gold_sql)
```

```python
gold_df = df[df["Medal"] == "Gold"].groupby("Team").size().sort_values(ascending=False).head(10)

print(gold_df)
```

2 F) Find out top 10 countries after number of medals

```python
top_medals_sql = query("""         
    SELECT Team, COUNT(Medal)
    FROM df
    WHERE Medal IS NOT NULL
    GROUP BY TEAM
    ORDER BY COUNT(Medal) DESC
    LIMIT 10
""")
top_medals_sql
    
```

```python
top_medals_df = df[df["Medal"].isin(["Gold", "Silver", "Bronze"])].groupby("Team").size().sort_values(ascending=False).head(10)

print(top_medals_df)

                        
                        
                                                             
```
